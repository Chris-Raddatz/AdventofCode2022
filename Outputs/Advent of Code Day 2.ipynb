```python
# A/X is Rock and gets one point
# B/Y is Paper and gets two points
# C/Z is Scissors and gets three points
# Get 0 points if you lose, 3 if it's a draw and 6 if you win
```

```python
inputs
```

```python
def points_won(list):
    sums = 0
    
    for i in list:
        if i[0] == 'A':
            if i[2] == 'Y':
                sums += 8
            elif i[2] == 'Z':
                sums += 3
            else:
                sums += 4
        if i[0] == 'B':
            if i[2] == 'X':
                sums += 1
            elif i[2] == 'Z':
                sums += 9
            else:
                sums += 5
        if i[0] == 'C':
            if i[2] == 'X':
                sums += 7
            elif i[2] == 'Y':
                sums += 2
            else:
                sums += 6
    return sums
```

```python
points_won(inputs)
```

```python
def determine_outcome(list):
    sums = 0
    
    for i in list:
        print(i)
        if i[2] == 'Z':
            sums += 6
            if i[0] == 'A':
                sums += 2
            if i[0] == 'B':
                sums += 3
            if i[0] == 'C':
                sums += 1
        if i[2] == 'Y':
            sums += 3
            if i[0] == 'A':
                sums += 1
            if i[0] == 'B':
                sums += 2
            if i[0] == 'C':
                sums += 3
        if i[2] == 'X':
            if i[0] == 'A':
                sums += 3
            if i[0] == 'B':
                sums += 1
            if i[0] == 'C':
                sums += 2
        print(sums)
    return sums
```
