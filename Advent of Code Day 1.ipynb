```python
with open(r"/Users/raddatzcd/Advent of Code/Elf Food.txt") as f:
    groups = f.read().split('\n\n')
```

```python
groups
```

```python
totals = list()

for e in groups:
    calories = 0
    for n in e.split():
        calories += int(n)
    totals.append(calories)
```

```python
totals.sort(reverse = True)
```

```python
print('The most calories is:', totals[0])
```

```python
print("The top three elves are carrying:", sum(totals[:3]))
```
